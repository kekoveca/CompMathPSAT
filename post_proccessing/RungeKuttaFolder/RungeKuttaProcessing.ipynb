```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style= 'whitegrid', font_scale=1.3)
```

```python
data1 = pd.read_csv('err.txt', header= None, dtype= np.float64)
data2 = pd.read_csv('err1.txt', header= None, dtype= np.float64)
h = [1 * 0.5 ** i for i in range(0, 13)]
x_data = np.log(h)
y_data1 = np.log(data1.iloc[0, :-1])
y_data2 = np.log(data2.iloc[0, :-1])
y_data1
```

```python
p3 = np.polyfit(x_data, y_data1, 1)
y_3 = np.polyval(p3, x_data)
k3 = p3[0]

p4 = np.polyfit(x_data, y_data2, 1)
y_4 = np.polyval(p4, x_data)
k4 = p4[0]


plt.figure(figsize= (16,10), dpi= 200)
plt.scatter(x_data, y_data1, c = 'red', label = f'$y\' = t^3$, ошибка порядка или меньше $10^{{-15}}$ на всем интервале, т.к. функция интегрируется этим методом точно')
plt.plot(x_data, y_data1, c = 'red', linestyle = '--', alpha = 0.3)
plt.scatter(x_data, y_data2, c = 'green', label = f'$y\'\' = - y, k = {k4:.2f}$')
plt.plot(x_data, y_4, c = 'green', linestyle = '--')

plt.plot([-9, 0], [np.log(1e-15), np.log(1e-15)], c = 'black', label = 'err = 1e-15')

plt.xlim(-9, 0)
plt.title('Численное интегрирование методом Рунге-Кутты 4')
plt.legend(loc = 'upper left')
plt.xlabel('$\ln h$')
plt.ylabel('$\ln err$')
plt.savefig(f'plot')
```
